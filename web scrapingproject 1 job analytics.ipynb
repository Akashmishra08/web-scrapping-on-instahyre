```python
# importing important libreries
from selenium import webdriver 
from selenium.webdriver.chrome.service import Service 
import time 
import pandas as pd
from bs4 import BeautifulSoup
import time
from datetime import datetime
```

```python
# website link
url = 'https://www.instahyre.com/python-jobs'

# creating varriable for storing the data 
name, location, founded, employee, about, skill, link = [], [], [], [], [], [], []

#Initialize webdriver (chrome) 
driver = webdriver.Chrome()

#open website
driver.get(url)

#wait for the page to load
time.sleep(8)

# loop through the pages
for i in range(1,20):
    #find element for each data type
    nam = driver.find_elements("xpath","//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']/a/div[@class='employer-job-name']")
    loc = driver.find_elements("xpath","//div[@class='employer-locations']/span/span[@class='ng-binding ng-scope']")
    foun = driver.find_elements("xpath","//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']//span[@ng-if='opp.employer.company_founded']")
    emp = driver.find_elements("xpath","//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']//span[@ng-if='opp.employer.employee_count']")
    abo = driver.find_elements("xpath","//div[@ng-if='opp.employer.instahyre_note']")
    ski = driver.find_elements("xpath","//ul[@class='tags candidate-opp-keywords']")
    lin = driver.find_elements("xpath","//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']/a")

    # Iterate over elements and append data to respective lists
    for n,l,f,e,a,li in zip(nam,loc,foun,emp,abo,lin):
       name.append(n.text)
       location.append(l.text)
       founded.append(f.text)
       employee.append(e.text)
       about.append(a.text)
       link1 = li.get_attribute("href")
       link.append(link1)
    soup = BeautifulSoup(driver.page_source,'html.parser')
    skill_block = soup.find_all('ul',class_= "tags candidate-opp-keywords")
    

    for s in skill_block:
        skl = []
        sk = s.find_all('li')

        for j in sk:
            skl.append(j.text)
            
        skill.append(skl)   
       
    #find the next page button and click it   
    button = driver.find_element("xpath","//li[@ng-click='nextPage()']") 
    button.click()

    # Wait for the next page to load
    time.sleep(8)
    
#close the web driver    
driver.quit()
```

```python
# print the list of all data
print(len(name))
print(len(location))
print(len(founded))
print(len(about))
print(len(skill))
print(len(link))
```
