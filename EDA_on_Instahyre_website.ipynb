<h4> Importing Pandas library and the dataset</h4>

```python
import pandas as pd
```

```python
df=pd.read_csv("insta_hire_data.csv")
df
```

```python
df.isnull().sum()
```

<h3> INSIGHT 1: Bar Plot for Frequency of number of employees</h3>

```python
import matplotlib.pyplot as plt
plt.figure()
df["employees"].value_counts().plot(kind='bar',color="green")
plt.show()
```

<h3>INSIGHT 2: Pie chart for Top 5 job positions </h3>

```python
#Frequecy of job position

a=df['job_position'].value_counts()
a
```

```python
# first extracting the right part from job positon column(sep="-")
a=df['right_part_job_position'] = df['job_position'].apply(lambda x: x.split('-')[-1])
a

a=df['right_part_job_position'].value_counts()
a.head()
plt.figure(figsize=(10, 7))
plt.pie(a.head().values, labels=a.head().index, autopct="%1.1f%%")
plt.title("Top 5 Job Positions")
plt.show()
```

<h3> INSIGHT 3: Top 10 Most demanding skills for Python Roles</h3>

```python
# insights from Skills column

# Extracting the top 10 skills from the "skills" column
top_10_skills = df['Skills'].str.split(',').explode().value_counts().head(10)
top_10_skills

# Creating a bar plot of the top 10 skills
plt.figure(figsize=(10, 10))
top_10_skills.plot(kind='bar',color="purple")
plt.xlabel("Skills", labelpad=15)
plt.ylabel("Count", labelpad=15)
plt.title("Count of top 10 skills")
plt.xticks(rotation=90)
plt.show()
```

<h3>INSIGHT 4: Pie chart for the Top Oppurtunities at different Job Locations</h3>

```python
# Top 10 locations with maximum Opputunities in Ascending order
a=df['right_part_job_location'] = df['job_location'].str.split('in').str[-1].head(10)
a
```

```python
a=df['right_part_job_location'] = df['job_location'].str.split('in').str[-1]
b=a.value_counts()
plt.figure(figsize=(10, 7))
plt.pie(b.values, labels=b.index, autopct="%1.1f%%")
plt.title("Job listings Oppurtunities at different Job Locations")
plt.show()
```

<h3>INSIGHT 5: Bar plot for number of job postings per establishment year</h3>

```python
# Bar plot for number of job postings per establishment year
plt.figure(figsize=(10, 5))
df['establishment_year'].value_counts().plot(kind='bar', color='orange')
plt.xlabel("Establishment Year", labelpad=15)
plt.ylabel("Count", labelpad=15)
plt.title("Number of Job Postings per Establishment Year")
plt.xticks(rotation=90)
plt.show()
```

<h1>FINAL INSIGHT: Companies founded in 2015 with job profiles of Data Scientist and Python Developer has the highest job listings at Bangalore location</h1>
